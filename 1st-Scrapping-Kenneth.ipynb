```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://mydramalist.com/shows/top'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
movietitle = soup.select('h6.title')
print(movietitle[0])
print(" ")
print("how many titles are available?",len(soup.select('h6.title')))
```

```python
text = soup.select('h6 a')
print(text)
```

```python
links=[]
for a in soup.select('h6 a'):
  links.append(a.attrs.get('href'))

print(links[0])
```

```python
movie_title=[]
links=[]
rating = []
type = []
year_of_release=[]
number_of_episodes= []

for t in soup.select('h6.title a'):
  if t.attrs.get('href')!= None:
    movie_title.append(t.get_text())
    links.append("https://mydramalist.com"+t.attrs.get('href') )

for h in soup.select('span.score'):
  rating.append(h.get_text())

for g in soup.select('span.text-muted'):
  types = g.get_text().split("-")
  type.append(types[0])
  years = g.get_text().split(" ")
  year_of_release.append(years[3])
  episode = g.get_text().split(",")
  number_of_episodes.append(episode[1])



print(movie_title)
print(links)
print(rating)
print(type)
print(year_of_release)
print(number_of_episodes)
```

```python
df = pd.DataFrame(
    {'drama title': movie_title,
     'drama review link': links,'ratings':rating, 'Drama Type':type, "Year Of release":year_of_release, "Number of episodes":number_of_episodes}
    )

print (df.head())

df.to_csv('MyDramaList.csv', index=False)
pd.read_csv('MyDramaList.csv')
```
