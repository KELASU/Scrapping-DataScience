```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://myanimelist.net/topmanga.php'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
movietitle = soup.select('h3.manga_h3')
print(movietitle[0])
print(" ")
print("how many titles are available?",len(soup.select('h3.manga_h3')))
```

```python
text = soup.select('h3 a')
print(text)
```

```python
links=[]
for a in soup.select('h3 a'):
  links.append(a.attrs.get('href'))

print(links[0])
```

```python
manga_title=[]
links=[]
rating = []
dates=[]
number_of_volumes= []
members =[]

for t in soup.select('h3.manga_h3 a'):
  if t.attrs.get('href')!= None:
    manga_title.append(t.get_text())
    links.append("https://mydramalist.com"+t.attrs.get('href') )

for b in soup.select('div.information'):
  data = b.get_text().split('\n')
  dates.append(data[2])
  number_of_volumes.append(data[1])
  members.append(data[3])

for h in soup.select('span.score-label'):
  if h.get_text()!='N/A':
    rating.append(h.get_text())

print(manga_title)
print(links)
print(dates)
print(number_of_volumes)
print(members)
print(rating)
```

```python
df = pd.DataFrame(
    {'Manga title': manga_title,
     ' link': links,'date':dates,'Volume':number_of_volumes,'Ratings':rating,'Members':members}
    )

print (df.head())

df.to_csv('MyMangaList.csv', index=False)
pd.read_csv('MyMangaList.csv')
```
